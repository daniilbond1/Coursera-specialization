```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats 
import math
```

Сгенерируем выборку объёма 1000 из бета распределенния с парметрами a=0.5 и b=2

```python
beta = scipy.stats.beta(0.5,2)
sample = beta.rvs(1000)
```

Построим графики гистограммы выборки и теоретической плотности(Почему-то не работала нормировка, так и не понял почему поэтому раскидал по двум графикам)

```python
x = np.linspace(0,2,1000)
cdf=beta.pdf(x)

plt.plot(x, cdf, label='theoretical PDF')
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
plt.hist(sample, bins =50)
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
```

Оценим распределение выборочного среднего случайной величины при разных объёмах выборок(5,10,50)


```python
#напишем функцию возвращающую масив средних по выборке объема n

def middle(n):
    sample=np.array([])
    i=1
    while i <= 1000:
        sample2 = beta.rvs(n)
        average=sum(sample2)/n
        sample=np.append(sample,average)
        i+=1
    return sample
 
```

Построим гистограмму для выборки при n=5

```python
plt.hist(middle(5), bins =150)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

Построим гистограмму для выборки при n=10

```python
plt.hist(middle(10), bins =150)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

Построим гистограмму для выборки при n=50

```python
plt.hist(middle(50), bins =150)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

Посчитаем теоретические мат.ожидание и дисперсию для нашего распределения и построим его график(по формулам из Википедии)

```python
mu=0.5/2.5
dispersion=(1)/((2.5**2)*3.5)
sigma=math.sqrt(dispersion)

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, scipy.stats.norm.pdf(x, mu, sigma))
plt.show()
```
