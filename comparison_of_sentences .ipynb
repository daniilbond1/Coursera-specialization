```python
import string
import numpy as np
import scipy
import scipy.spatial
```

```python
file=open('sentences.txt','r')
Mylist = file.read().split("\n")[:]
e = ",", ".","2","0","1","3","4","5","6","7","8","9","(",")","/","\"",">","<"," - ","\'"
```

```python
for i in range (len(Mylist)):
    Mylist[i]=Mylist[i].lower()
    for m in range(len(e)):
     Mylist[i]= Mylist[i].replace(e[m], " ")
```

```python
for i in range (len(Mylist)):
    Mylist[i]=Mylist[i].split(" ")[:-1]
    
    my_dict  = {}
    z=0
    
for i in range (len(Mylist)):
    for m in range(len(Mylist[i])):
        key_exists = Mylist[i][m] in my_dict 
        if ((key_exists==False) & (Mylist[i][m]!='')):
         my_dict[Mylist[i][m]]=z
         z+=1    
```

```python
my_dict
```

```python
My_matrix=np.zeros((22,253))

for i in range (len(Mylist)):
    for m in range(len(Mylist[i])):
        if Mylist[i][m]!='':
         My_matrix[i,my_dict[Mylist[i][m]]]+=1
```

```python
My_matrix
```

```python
answer = []
for row in My_matrix:
    answer.append(scipy.spatial.distance.cosine(My_matrix[0,:], row))
```
